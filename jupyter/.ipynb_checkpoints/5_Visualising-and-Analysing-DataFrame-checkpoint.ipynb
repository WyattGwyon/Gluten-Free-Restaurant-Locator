```python
pr = []
for p in gfdf["price_level"]:
    if p == 0:
        pr.append("Free")
    elif p == 1:
        pr.append("Inexpensive")
    elif p == 2:
        pr.append("Moderate")
    elif p == 3:
        pr.append("Expensive")
    elif p == 4:
        pr.append("Very Expensive")
    else:
        pr.append("Unknown")
    
```

```python
m_layer2 = Map(location=[40.4169,-3.7035], zoom_start=16)
fg = FeatureGroup(name="unique")

fg_uni_rest = FeatureGroup(name = "Unique Restaurants")
fg_uni_bake = FeatureGroup(name = "Unique Bakery")
fg_chain_rest = FeatureGroup(name = "Chain Restaurants")
fg_chain_bake = FeatureGroup(name = "Chain Bakery")
fg_locale = FeatureGroup(name = "Gluten Free Locale")

for i , row in gfdf.iterrows():
    rest = {
        "location":row["geometry"],
        "popup":row["name"],
        "tooltip":row["name"]
    }
    
    if row["chain"]=="chain" and row["types"]=="restaurant":
        ic = Icon(color="blue",
                  prefix="fa",
                  icon="cutlery",
                  icon_color= "black")
        Marker(**rest, icon=ic).add_to(fg_chain_rest)
    elif row["chain"]=="chain" and row["types"]=="bakery":
        ic = Icon(color="blue",
                  prefix="fa",
                  icon = "bread-slice",
                  icon_color="black")
        Marker(**rest, icon=ic).add_to(fg_chain_bake)
    elif row["chain"]=="unique" and row["types"]=="bakery":
        ic = Icon(color="red",
                  prefix="fa",
                  icon = "bread-slice",
                  icon_color="white")
        Marker(**rest, icon=ic).add_to(fg_uni_bake)
    elif row["chain"]=="unique" and row["types"]=="restaurant":
        ic = Icon(color="red",
                  prefix="fa",
                  icon = "cutlery",
                  icon_color="white")
        Marker(**rest, icon=ic).add_to(fg_uni_rest)
    
    else:
        ic = Icon(color="green",
                  prefix="fa",
                  icon="thumbs-up",
                 icon_color="white")
        Marker(**rest, icon=ic).add_to(fg_locale)
        
fg_uni_rest.add_to(m_layer2)
fg_uni_bake.add_to(m_layer2)
fg_chain_rest.add_to(m_layer2)
fg_chain_rest.add_to(m_layer2)
fg_locale.add_to(m_layer2)

Choropleth(
    #Geo Data - Border
    geo_data="data/madrid.geojson.json",
    key_on="features.properties.name",
    name="Barrios",
    fill_opacity=0.1,
    line_opacity=2,
    line_weight= 1,
    highlight=True,
    smooth_factor=2
    #data=gfdf,
    #columns=['name', 'user_ratings_total'],
    #legend_name='Unemployment Rate (%)',
    
).add_to(m_layer2)

        
LayerControl(collapsed=False).add_to(m_layer2)
        
m_layer2
```

# Chart Examples to use as Templates

```python
df = px.data.gapminder().query("year == 2007")
fig = px.sunburst(df, path=['continent', 'country'], values='pop',
                  color='lifeExp', hover_data=['iso_alpha'],
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint=np.average(df['lifeExp'], weights=df['pop']))
fig.show()
```

```python
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']

fig = go.Figure(data=[go.Pie(labels=['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen'],
                             values=[4500,2500,1053,500])])
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                  marker=dict(colors=colors, line=dict(color='#000000', width=2)))
fig.show()
```

```python
import plotly.express as px
df = px.data.wind()
fig = px.bar_polar(df, r="frequency", theta="direction",
                   color="strength", template="plotly_dark",
                   color_discrete_sequence= px.colors.sequential.Plasma_r)
fig.show()
```
